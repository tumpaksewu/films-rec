```python
import pandas as pd 
import numpy as np
```

```python
df = pd.read_csv('./../data/films_all.csv')
```

```python
df.head(10)
```

```python
from sentence_transformers import SentenceTransformer

model = SentenceTransformer('sentence-transformers/paraphrase-multilingual-MiniLM-L12-v2')
```

```python
descriptions = df.description.values.tolist()

# 2. Calculate embeddings by calling model.encode()
embeddings = model.encode(descriptions)
```

```python
query = 'действие происходит на диком западе, ковбои и индейцы'
query_emb = model.encode(query)

# 3. Calculate the embedding similarities
similarities = model.similarity(query_emb, embeddings)

similarities
```

```python
n = 10
top_n_idx = np.argsort(similarities).squeeze(0)[-n:].tolist()

for n in top_n_idx:
    print(df.iloc[n]['title'])
    print(df.iloc[n]['description'])
    print('---')
```
